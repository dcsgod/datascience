```python
import pandas as pd
import numpy as np

# Define the number of samples in the dataset
num_samples = 1000

# Define the possible values for each feature
socioeconomic_status = ['Low', 'Medium', 'High']
age_range = range(18, 75)
gender = ['Male', 'Female']

# Create the dataset
data = pd.DataFrame(columns=['Socioeconomic_Status', 'Age', 'Gender', 'Survival'])

for _ in range(num_samples):
    socioeconomic = np.random.choice(socioeconomic_status)
    age = np.random.choice(age_range)
    gender_val = np.random.choice(gender)
    survival = np.random.choice([0, 1], p=[0.3, 0.7])  # Adjust the probabilities as desired

    data = data.append({
        'Socioeconomic_Status': socioeconomic,
        'Age': age,
        'Gender': gender_val,
        'Survival': survival
    }, ignore_index=True)

# Save the dataset to a CSV file
data.to_csv('survival_data.csv', index=False)
```
